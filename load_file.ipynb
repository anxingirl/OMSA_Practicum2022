## READ IN PHARMACY FILES THAT HAS BEEN EXTRACTED (don't do this 2nd time)

```python
pharmacy_dir = "/content/drive/MyDrive/Colab Notebooks/practicum_data/2022-09-04 17.10.52 - Claims Ananth Venugopal/"
pharmacy_path = "/content/drive/MyDrive/Colab Notebooks/practicum_data/2022-09-04 17.10.52 - Claims Ananth Venugopal/Pharmacy_Claims_extracted/"
```

```python
# list all the file box 

pharmacy = []

columns_stay = ['_c0','_c7','_c10','_c11','_c19','_c24','_c25','_c26','_c27','_c28','_c32','_c33','_c36','_c37','_c38','_c39','_c40','_c41']    

for m_dir in os.listdir(pharmacy_path):
  path = pharmacy_path + m_dir
  for f in os.listdir(path):
    #print(path + "/" + f)
    if f.endswith("parquet"):
      pharmacy.append(pd.read_parquet(path+"/"+f, columns = columns_stay))
    # read in specific columns
```

```python
pharmacy_combine = pd.concat(pharmacy)
pharmacy_combine = pharmacy_combine.reset_index(drop = True)
```

```python
columns_names = ['Member Life ID',
                 'Risk Type','Birth Date','Gender Code',
 'New Prescription Indicator',
 'Prescribing Provider DEA ID',
 'Prescribing Provider Name',
 'NDC',
 'Drug Name',
 'Specialty Drug Indicator',
 'Therapeutic Class',
 'Strength',
 'Fill Date',
 'Paid Date',
 'Days Supply',
 'Quantity Dispensed',
 'Refill Number',
 'Number Of Refills'
]
pharmacy_combine.columns = columns_names
```

```python
pharmacy_combine.to_csv("/content/drive/MyDrive/Colab Notebooks/practicum_data/pharmacy_combine_total.csv",index=False)
```
